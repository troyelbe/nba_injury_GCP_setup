```python
#GCP configs needed
from google.cloud import storage
```

```python
today = datetime.today()

year = today.strftime("%Y")
month = today.strftime("%b")
day = today.strftime("%d")
```

```python

project_id = <"YOUR_PROJECT_ID">
bucket_name = <"YOUR_BUCKET_NAME">

local_filename = '/content/drive/MyDrive/injury_report_master.csv'
gcs_path = f'raw/injuries/{year}/{month}/{day}/injuries.csv'

client = storage.Client()
bucket = client.bucket(bucket_name)
blob = bucket.blob(gcs_path)

blob.upload_from_filename(local_filename)

print(f"Uploaded to gs://{bucket_name}/{gcs_path}")
```

```python
from google.cloud import bigquery

project_id = <"YOUR_PROJECT_ID">
dataset_id = <"YOUR_DATASET">
table_id = <"YOUR_TABLE">


gcs_uri = f"gs://nba-injury-data-2026/raw/injuries/{year}/{month}/{day}/injuries.csv"

client = bigquery.Client(project=project_id)

table_ref = f"{project_id}.{dataset_id}.{table_id}"

job_config = bigquery.LoadJobConfig(
    source_format=bigquery.SourceFormat.CSV,
    skip_leading_rows=1,
    autodetect=True,
    write_disposition="WRITE_APPEND"
)

load_job = client.load_table_from_uri(
    gcs_uri,
    table_ref,
    job_config=job_config
)

load_job.result()

print("Load complete.")
```
